```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('HR-Employee-Attrition.csv')
```

```python
total_employees=len(df)
total_employees
```

```python
plt.figure(figsize=(5,7))
splot=sns.countplot(df['Attrition'])
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.title('Attrition')
plt.show()
```

```python
plt.figure(figsize=(7,7))
splot=sns.countplot('Department',data=df,hue='Attrition',palette='pastel')
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

plt.title('Attrition per Department')
plt.show()
```

```python
df_attrition_yes=df[df['Attrition']=='Yes']
```

```python
plt.figure(figsize=(20,10))
sns.countplot('Attrition',data=df_attrition_yes,hue='TotalWorkingYears')
plt.title('Attrition based on the number of years')
```

```python
plt.figure(figsize=(20,10))
sns.countplot('Attrition',data=df_attrition_yes,hue='Department')
plt.title('Attrition based on the number of years')
```

```python
df_att_count_yes= df_attrition_yes.groupby(['Department','TotalWorkingYears']).count()   
```

```python
df_att_count_yes['Percentage']=df_att_count_yes['Attrition']/len(df_att_count_yes)*100
```
