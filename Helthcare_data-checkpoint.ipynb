```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("Healthcare_Investments_and_Hospital_Stay (1).csv")
print(df.head())
```

```python
df_loc=set(df['Location'])
print(df_loc)
```

```python
plt.figure(figsize=(20,10))
g=sns.countplot(df['Location'],orient="h")
plt.xticks(rotation=30)
plt.title("Location Wise Count")
plt.show()
```

```python
print("Minimum Year",min(df['Time']))
print("Maximum Year",max(df['Time']))
```

```python
plt.figure(figsize=(10,10))
g=sns.distplot(df['Time'],kde=False)
plt.title("Time Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df['Hospital_Stay'],kde=False)
plt.title("Distribution of Hospital Stay")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.distplot(df["MRI_Units"],kde=False)
plt.title("Distribution of MRI Units")
plt.show()
```
